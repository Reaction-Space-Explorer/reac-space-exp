```python
from ambit_tautomer import generateTautomers
import pandas as pd

# Using rdkit for the purpose of drawing, not generating tautomers
from rdkit.Chem import Draw, MolFromSmiles
from rdkit.Chem.Draw import IPythonConsole
```

Please see if Ambit, much like RDKit, goes too far and creates molecules that aren't even tautomers. These are a few product molecules randomly taken from the glucose hydrolysis reaction network.

```python
test_parents = ['C1(CO)OC(C(O)=CC=1)O', 'C1=CC(CO)OC(C=1O)O', 'C1=C(O)OC(CO)C(C=1)O', 'C1(OC(C(O)=CC=1)CO)O', 'C1=C(C(CO)OC(C=1)O)O']
parent_mols = [MolFromSmiles(mol) for mol in test_parents]
print("These are the molecules I'm using to generate the tautomers")
Draw.MolsToGridImage(parent_mols, molsPerRow=4, subImgSize=(200,200))
```

Apparently, for loop doesn't work for Draw.MolsToGridImage. But anyways

```python
tautomer_mols = []
for smiles in test_parents:
    tautomers = generateTautomers(smiles, 'IA-DFS') # second argument is the algorithm to use
    tautomer_mols.append([MolFromSmiles(taut) for taut in tautomers])

print('These are the tautomers due to the parent molecules, each set is separated in separate cells')
Draw.MolsToGridImage(tautomer_mols[0], molsPerRow=6)
```

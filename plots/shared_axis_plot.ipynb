```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("YSP_ReactionNetwork_Descriptor_Data_November2021.csv")
sns.set_context("paper", font_scale=2.2);
sns.set_style("ticks");
fig, axes = plt.subplots(1,3, figsize=(28,20), sharex=True);
reversed_data = df.iloc[::-1]
sns.scatterplot(ax=axes[0], data=reversed_data, x='Exact Mass',y='Number of Carbons', hue="Generation", palette="hls", marker="o", size="Generation", sizes=(130, 60), alpha=.98, legend="full")
sns.scatterplot(ax=axes[1], data=reversed_data, x='Exact Mass', y='Number of Oxygens', hue="Generation", palette="hls", marker="o", size="Generation", sizes=(130, 60), alpha=.98, legend="full")
sns.scatterplot(ax=axes[2], data=reversed_data, x="Exact Mass", y='Number of Hydrogens', hue="Generation", palette="hls", marker="o", size="Generation", sizes=(130, 60), alpha=.98, legend="full")
```

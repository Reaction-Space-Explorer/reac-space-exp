```python
import matplotlib.pyplot as plt
import pandas as pd
my_csv = pd.read_csv('dbmatch.csv', sep=',')
print(my_csv.columns)
x_values = my_csv['Generation']
fig = plt.figure(figsize=(6.5,6.5))
plt.plot(x_values, my_csv['\tComputed Set'], 'ro-', label='Computed Set',linewidth=3.5, marker='o', markerfacecolor='black', markeredgewidth=2,
         markersize=8,)
plt.plot(x_values, my_csv['HMDB'], 'bo-', label='HMDB',marker='o', linewidth=1.5, markerfacecolor='black', markeredgewidth=2,
         markersize=7,)
plt.plot(x_values, my_csv['KEGG'], 'go-', label='KEGG',marker='o', linewidth=1.5, markerfacecolor='black', markeredgewidth=2,
         markersize=7,)
plt.plot(x_values, my_csv['ECMDB'], 'mo-', label='ECMDB',marker='o', linewidth=1.5, markerfacecolor='black', markeredgewidth=2,
         markersize=7,)
plt.legend(loc='upper left', bbox_to_anchor=(0.02, 0.99), fontsize=11)
plt.ylabel('Number of molecules ($log_{10}$)', fontsize=14)
plt.xlabel('Generation', fontsize=14)
plt.yscale('log')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

my_csv = pd.read_csv('https://raw.githubusercontent.com/Reaction-Space-Explorer/reac-space-exp/master/plots/database_match.csv.csv', sep=',')

print(my_csv.columns)

x_values = my_csv['Generation']

fig = plt.figure(figsize=(6.5,6.5))

plt.plot(x_values, my_csv['\tComputed Set'], label='Computed Set', linestyle='--', mfc='magenta', marker='^',
         color='darkgray', mec='gray', linewidth=1.8,#markeredgewidth=2,  ,
         markersize=10)
plt.plot(x_values, my_csv['HMDB'], color='silver', mfc='dodgerblue', label='HMDB',marker='s', linewidth=1.7, mec='gray', #markeredgewidth=2,
         markersize=8,
        )
plt.plot(x_values, my_csv['KEGG'], color='grey', linestyle='--', label='KEGG', marker='o', mfc='gold',
         linewidth=1.7, mec='silver', #markeredgewidth=2,
         markersize=8,)
plt.plot(x_values, my_csv['ECMDB'], color='dimgray', mfc='crimson', linestyle='-.', label='ECMDB', marker='*', linewidth=1.7,
         mec='darkgray', #markeredgewidth=2,
         markersize=8,)
plt.legend(loc='upper left', bbox_to_anchor=(0.02, 0.99), fontsize=13)
plt.ylabel('Number of Molecules', fontsize=13)
plt.xlabel('Generation', fontsize=13)
plt.yscale('log')
plt.savefig('database_comparison.png', dpi=600)
plt.show()
```

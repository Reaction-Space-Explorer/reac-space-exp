```python
import matplotlib.pyplot as plt
import pandas as pd
my_csv = pd.read_csv('database_match.csv', sep=',')
print(my_csv.columns)
x_values = my_csv['Generation']
fig = plt.figure(figsize=(6.5,6.5))
plt.plot(x_values, my_csv['Computed Set'], 'go-', label='Computed Set',linewidth=4,marker='o', markerfacecolor='black', markeredgewidth=2,
         markersize=8,)
plt.plot(x_values, my_csv['KEGG'], 'mo-', label='KEGG',marker='o', markerfacecolor='black', markeredgewidth=2,
         markersize=8,)
plt.plot(x_values, my_csv['HMDB'], 'co-', label='HMDB',marker='o', markerfacecolor='black', markeredgewidth=2,
         markersize=8,)
plt.legend(loc='upper left', bbox_to_anchor=(0.02, 0.99), fontsize=11)
plt.ylabel('Number of molecules (Log10)', fontsize=14)
plt.xlabel('Generation', fontsize=14)
plt.yscale('log')
plt.show()
```

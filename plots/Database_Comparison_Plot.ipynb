```python
import matplotlib.pyplot as plt
import pandas as pd
my_csv = pd.read_csv('dbmatch.csv', sep=',')
print(my_csv.columns)
x_values = my_csv['Generation']
fig = plt.figure(figsize=(6.5,6.5))
plt.plot(x_values, my_csv['\tComputed Set'], 'ro-', label='Computed Set',linewidth=3.5, marker='o', markerfacecolor='black', markeredgewidth=2,
         markersize=8,)
plt.plot(x_values, my_csv['HMDB'], 'bo-', label='HMDB',marker='o', linewidth=1.5, markerfacecolor='black', markeredgewidth=2,
         markersize=7,)
plt.plot(x_values, my_csv['KEGG'], 'go-', label='KEGG',marker='o', linewidth=1.5, markerfacecolor='black', markeredgewidth=2,
         markersize=7,)
plt.plot(x_values, my_csv['ECMDB'], 'mo-', label='ECMDB',marker='o', linewidth=1.5, markerfacecolor='black', markeredgewidth=2,
         markersize=7,)
plt.legend(loc='upper left', bbox_to_anchor=(0.02, 0.99), fontsize=11)
plt.ylabel('Number of molecules [$log_{10}$]', fontsize=14)
plt.xlabel('Generation', fontsize=14)
plt.yscale('log')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

my_csv = pd.read_csv('database_match.csv.csv', sep=',')

print(my_csv.columns)

x_values = my_csv['Generation']

fig = plt.figure(figsize=(6.5,6.5))

plt.plot(x_values, my_csv['\tComputed Set'], label='Computed Set', linestyle='--', mfc='magenta', marker='^',
         color='darkgray', mec='gray', linewidth=1.5,#markeredgewidth=2,  ,
         markersize=8)
plt.plot(x_values, my_csv['HMDB'], color='silver', mfc='dodgerblue', label='HMDB',marker='s', linewidth=1.5, mec='gray', #markeredgewidth=2,
         markersize=7,
        )
plt.plot(x_values, my_csv['KEGG'], color='grey', linestyle='--', label='KEGG', marker='o', mfc='gold',
         linewidth=1.5, mec='silver', #markeredgewidth=2,
         markersize=7,)
plt.plot(x_values, my_csv['ECMDB'], color='dimgray', mfc='crimson', linestyle='-.', label='ECMDB', marker='*', linewidth=1.5,
         mec='darkgray', #markeredgewidth=2,
         markersize=9,)
plt.legend(loc='upper left', bbox_to_anchor=(0.02, 0.99), fontsize=12)
plt.ylabel('Number of molecules ($log_{10}$)', fontsize=13)
plt.xlabel('Generation', fontsize=13)
plt.yscale('log')
plt.savefig('database_comparison.jpg', dpi=300)
plt.show()
```
